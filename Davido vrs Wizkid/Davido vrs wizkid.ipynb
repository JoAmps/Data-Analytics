```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
music=pd.read_csv('nigerian_spotify_songs1.csv')
music.head(3)
```

```python
music.shape
```

```python
#missing values
music.isnull().sum()
```

```python
music['artist'].value_counts()
```

```python
#only use data from davido and wizkid
musicians=music[(music['artist']=='DaVido') | (music['artist']=='WizKid')]
musicians.head(20)
```

```python
musicians['artist'].value_counts()
```

```python
print(musicians.groupby('artist').agg('popularity').mean());

fig1, ax1 = plt.subplots(figsize=(7,7))
labels=['Davido','Wizkid']
explode=(0,0.1)
plt.pie(musicians.groupby('artist').agg('popularity').mean(),labels=labels, explode=explode,shadow=True,
        startangle=90,autopct='%.2f')
ax1.axis('equal');
ax1.set_title('Most Popular Artist',fontdict={'size':20});
```

#### Wizkid had an average popularity score of 40.6 whiles davido had 32.9, which shows wizkid is rated more popular by spotify users than davido and hence he is the most popular

```python
musicians.groupby(['name','artist'])['popularity'].mean()
```

```python
ax=musicians.groupby(['name','artist'])['popularity'].mean().unstack().sort_values(by=['WizKid','DaVido'],ascending=False).plot(kind='barh',
                                                                                                     stacked=True, figsize=(7,7))
ax.set_title('Most Popular song',fontdict={'size':20});
```

#### Wizkid's Joro is his most popular song with a popularity rating of 66 whiles Davido's Blow my mind is his most popular song with a rating of 61, which means wizkid has the most popular song.

```python
print(musicians.groupby('artist').agg('danceability').mean())
```

```python
print(musicians.groupby('artist').agg('danceability').mean())
ax=musicians.groupby('artist').agg('danceability').mean().plot(kind='bar', figsize=(6,6))
ax.set_title('King of Dance Music',fontdict={'size':20});
```

#### The closer the value is to 1, the ease it is to be danced to. Davido's songs had a danceable rating of 0.734 compared to wizkid's songs whihc had a rating of 0.713, showing davido's songs are more danceable than wizkid's, crowning Davido the king of dance music over wizkid

```python
print(musicians.groupby('artist')['instrumentalness'].mean())
ax=musicians.groupby('artist')['instrumentalness'].mean().plot(kind='barh', color='green', figsize=(7,7))
ax.set_title('Best Instrumentals',fontdict={'size':20});
```

#### The closer the instrumentalness value is to 1.0, the better the instrumentals in the song. Wizkid's song has an average instrumental's score of 0.000419 whiles Davido's has 0.000057, therefore Wizkid's songs have the best instrumentals
